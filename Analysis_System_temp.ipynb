```python
import tkinter as tk
from tkinter import ttk
from tkinter import PhotoImage
from PIL import Image, ImageTk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import numpy as np

# 新建主視窗 (程式名、大小、Logo、背景顏色)
root = tk.Tk()
root.title("ITRI Analysis System")
root.geometry("1920x1080")   #1440, 1280
root.configure(bg='#7B7B7B') 
root.iconbitmap('itri.ico') 


current_page = None

def show_frame(frame):
    global current_page
    if current_page is not None:
        current_page.pack_forget()
    frame.pack(fill='both', expand=True)
    frame.tkraise()
    current_page = frame

sidebar_width = 256
main_window_width = 1920 - sidebar_width
main_window_height = 1080

Page_1 = tk.Frame(root, bg='#ff9999', width=main_window_width, height=main_window_height)
Page_1.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_2 = tk.Frame(root, bg='#99ff99', width=main_window_width, height=main_window_height)
Page_2.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_3 = tk.Frame(root, bg='#9999ff', width=main_window_width, height=main_window_height)
Page_3.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_4 = tk.Frame(root, bg='#ffff99', width=main_window_width, height=main_window_height)
Page_4.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_5 = tk.Frame(root, bg='#ff99ff', width=main_window_width, height=main_window_height)
Page_5.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_6 = tk.Frame(root, bg='#99ffff', width=main_window_width, height=main_window_height)
Page_6.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_7 = tk.Frame(root, bg='#9ffff9', width=main_window_width, height=main_window_height)
Page_7.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
Page_8 = tk.Frame(root, bg='#9ffff9', width=main_window_width, height=main_window_height)
Page_8.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

# 側邊框架
sidebar_frame = tk.Frame(root, bg='#20212c', width=256, height=1024)
sidebar_frame.pack(side=tk.LEFT, fill=tk.Y)

# 側邊欄按鈕
sidebar_buttons = ["Overview", "Chart", "Transactions", "Wallet", "News", "Mail Box", "Setting", "Logout"]

button_positions = [40, 104, 168, 236, 296, 360, 888, 952]  # Y-coordinates for buttons

# for i, button in enumerate(sidebar_buttons):                                              
#     tk.Button(sidebar_frame, text=button, fg='white', bg='#20212c', relief=tk.FLAT, font=('Arial',15,'bold')).place(x=40, y=button_positions[i])


pages = [Page_1, Page_2, Page_3, Page_4, Page_5, Page_6, Page_7, Page_8]

for page in pages:
    page.pack(fill='both', expand=True)
    page.pack_forget()



for i, (button, frame) in enumerate(zip(sidebar_buttons, pages)):
    action = lambda f=frame: show_frame(f)
    tk.Button(sidebar_frame, text=button, fg='white', bg='#20212c', relief=tk.FLAT,
              font=('Arial',15,'bold'), command=action).place(x=40, y=button_positions[i])



for frame in pages:
    frame.pack_forget()


    
# 上方框架
topbar_frame = tk.Frame(root, bg='#20212c', width=1000, height=50)
topbar_frame.pack(side=tk.TOP, fill=tk.X)

# 搜尋欄
search_bar = tk.Entry(topbar_frame, bg='#404040', fg='white')
search_bar.pack(side=tk.LEFT, padx=20, pady=10)

# 主要內容框架
content_frame = tk.Frame(root, bg='#1e1e2e')
content_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)


# 定義功能框架
frame_width = 258
frame_height = 178
frame_color = '#20212c'
frame_positions = [(316, 140), (634, 140), (952, 140), (1270, 140), (1590, 140)]

for pos in frame_positions:
    frame = tk.Frame(root, bg=frame_color, width=frame_width, height=frame_height, relief=tk.RAISED)
    frame.place(x=pos[0], y=pos[1])

# 定義功能框架內的小圖片

png_paths = [r'C:\Users\user\Desktop\Analysis_System_GUI-main\RMS.png',
              r'C:\Users\user\Desktop\Analysis_System_GUI-main\Peak.png',
              r'C:\Users\user\Desktop\Analysis_System_GUI-main\FFT.png',
              r'C:\Users\user\Desktop\Analysis_System_GUI-main\factor.png',
              r'C:\Users\user\Desktop\Analysis_System_GUI-main\Analysis.png']



# Positions for each png and label
positions = [(336, 165), (654, 165), (972, 165), (1290, 165), (1610, 165)]
label_texts = ["RMS", "Peak", "FFT", "Crest Factor", "Main Frequency"]
label_positions = [(396, 165), (714, 165), (1032, 165), (1350, 165), (1670, 165)]


# Dictionary to keep a reference to the images to prevent garbage collection
photo_images = {}

# Load, resize, and place each icon

for path, pos, label_text, label_pos in zip(png_paths, positions, label_texts, label_positions):
    # Load the image
    original_image = Image.open(path).convert('RGBA')  # Ensure image is in RGBA mode
    # Resize the image
    resized_image = original_image.resize((45, 45), Image.ANTIALIAS)
    # Convert the image for Tkinter
    photo_image = ImageTk.PhotoImage(resized_image)
    # Create a label and place the image in the UI
    image_label = tk.Label(root, image=photo_image, bg='#20212c')  # Set label background to match the window
    image_label.place(x=pos[0], y=pos[1])
    # Store the image reference
    photo_images[path] = photo_image

    # Create and place the text label
    text_label = tk.Label(root, text=label_text, bg=frame_color, fg='white', font=('Arial',15,'bold'))
    text_label.place(x=label_pos[0], y=label_pos[1]+10)         
        

        
chart_positions = [(396, 211), (714, 211), (1032, 211), (1350, 211), (1670, 211)]
chart_width, chart_height = 165, 100

for pos in chart_positions:
    # 創建一個圖形和軸，並設置背景透明
    fig = Figure(figsize=(chart_width/100, chart_height/100), dpi=100, facecolor='#20212c')

    ax = fig.add_subplot(111)
    ax.patch.set_facecolor('#20212c')  # 也為軸設置透明背景  
    ax.set_facecolor('#20212c')  # 圖表的背景顏色
    ax.spines['bottom'].set_color('white')  # X軸的顏色
    ax.spines['top'].set_color('white')
    ax.spines['right'].set_color('white')  # Y軸的顏色
    ax.spines['left'].set_color('white')

    ax.tick_params(axis='x', colors='white')  # X軸刻度線和刻度標籤的顏色
    ax.tick_params(axis='y', colors='white')  # Y軸刻度線和刻度標籤的顏色
       
    
    # 在這裡創建折線圖的數據 (示例)
    x = np.linspace(0, 2 * np.pi, 50)
    y = np.sin(x)
    ax.plot(x, y)

    # 將圖形加入到Tkinter窗口
    canvas = FigureCanvasTkAgg(fig, master=root)  # 'root' 是 Tkinter 主窗口
    canvas_widget = canvas.get_tk_widget()
    canvas_widget.place(x=pos[0], y=pos[1])    


    
        
# Logo 置放處

original_image = Image.open(r'C:\Users\user\Desktop\Analysis_System_GUI-main\Logo.png')  # Replace with your image file path
resized_image = original_image.resize((170, 41), Image.ANTIALIAS)
photo_image = ImageTk.PhotoImage(resized_image)

image_label = tk.Label(root, image=photo_image)
image_label.place(x=40, y=32) 
            
        
# 即時顯示

Realtime = tk.Frame(root, bg='#20212c', width=1533, height=634, relief=tk.RAISED)
Realtime.place(x=316, y=350)
   
# 在這裡創建折線圖的數據 (示例)
dpi = 100
width_inch = Realtime['width'] / dpi
height_inch = Realtime['height'] / dpi


fig = Figure(figsize=(width_inch, height_inch), dpi=dpi, facecolor='#20212c')
ax = fig.add_subplot(111)


ax.set_facecolor('#20212c')  # 圖表的背景顏色
ax.patch.set_facecolor('#20212c')  # 也為軸設置透明背景

ax.spines['bottom'].set_color('white')  # X軸的顏色
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')  # Y軸的顏色
ax.spines['left'].set_color('white')

ax.tick_params(axis='x', colors='white')  # X軸刻度線和刻度標籤的顏色
ax.tick_params(axis='y', colors='white')  # Y軸刻度線和刻度標籤的顏色



ax.plot(x, y)



# 將圖形加入到Tkinter窗口
canvas = FigureCanvasTkAgg(fig, master=Realtime)  # 'root' 是 Tkinter 主窗口
canvas_widget = canvas.get_tk_widget()
canvas_widget.pack(side=tk.TOP, fill=tk.BOTH, expand=True)
    




# 暫定內容區域
#tk.Label(content_frame, text="Main Content Area", bg='#1e1e2e', fg='white').pack(padx=20, pady=20)

root.mainloop()
```
